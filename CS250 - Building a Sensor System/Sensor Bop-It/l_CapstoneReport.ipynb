# Description

Our project is a recreation of the handheld game "bop-it", where the player has to perform specific actions (activate specific sensors) based on the game's commands. The game speeds up until the player loses.

# Hardware

Sensors used to play:

-Microphone

-Potentiometer

-Accelerometer

-Button

4 LEDs with resistors to provide information output

1 ADC to get information from sensors

# Operation

By performing the correct action/activating the correct sensors, the player gains a point. For every point the player gains in a row, the time limit given to perform each subsequent action decreases, until the player fails to perform the action, and the game ends. Game commands and information are given through flashing the LEDs.

Blue LED: Shout At It! - Microphone

Yellow LED: Twist It! - Potentiometer

Green LED: Shake It! - Accelerometer

Red LED: Bop It! - Button

After running the code, press the button to begin the game.
